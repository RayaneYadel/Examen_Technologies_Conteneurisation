```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

df = pd.read_csv('../data/city_lifestyle_dataset.csv')

X = df.drop(['city_name', 'country'], axis=1)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(df_pca['PC1'], df_pca['PC2'], alpha=0.6, edgecolors='w')
plt.title('Visualisation 2D des profils de villes (PCA)')
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.1%} de variance)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.1%} de variance)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

### Analyse de la structure des profils urbains (PCA)

L'examen du graphique 2D révèle une organisation cohérente des 300 villes du dataset. Avec près de *80% de la variance totale expliquée* (53.8% sur PC1 et 25.8% sur PC2), la réduction de dimension est particulièrement performante sur ce jeu de données.

#### Observations clés :

* *Axe Majeur (PC1 - 53.8%)* : Cet axe horizontal représente le facteur de différenciation principal. Au vu des corrélations du dataset, il semble opposer les *villes à haut revenu et forte infrastructure digitale* (à droite) aux *zones urbaines en développement* (à gauche).
* *Axe Secondaire (PC2 - 25.8%)* : L'axe vertical semble capturer la dimension *"Qualité de Vie vs Densité"*. Les points situés en haut de l'axe représentent probablement des villes avec une forte densité ou des enjeux de pollution marqués, tandis que le bas du graphique regroupe des profils plus "verts".
* *Structure des Clusters* : 
    * On observe une *forte concentration* dans la partie centrale droite, suggérant une homogénéité parmi les métropoles des régions développées (Europe/Amérique du Nord).
    * La dispersion plus importante sur la gauche indique une plus grande diversité de profils parmi les villes à revenus plus faibles ou intermédiaires.

#### 💡 Conclusion :
La PCA a permis de simplifier 10 indicateurs complexes en un plan 2D lisible sans perte majeure d'information. Cette base de données projetée est maintenant prête à être exportée pour calculer la métrique de *Trustworthiness* et être comparée aux méthodes non-linéaires (t-SNE/UMAP).
